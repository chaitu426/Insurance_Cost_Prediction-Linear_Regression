### Data encoding 
#### as many data col in dataset are catoguric and in string format but computer undestand the numeric data well so in this part we can convert strings --> numeric data 
#### for eg we have male & female so what we can do male --> 0 & female --> 1

```python
# encoding the sex col
insurence_data.replace({'sex':{'male':0, 'female':1}}, inplace=True)

# encoding the smoker col
insurence_data.replace({'smoker':{'yes':1, 'no':0}}, inplace=True)

# encoding the region col
insurence_data.replace({'region':{'southeast':0, 'southwest':1, 'northeast':2, 'northwest':3 }}, inplace=True)
```

### spliting the independent fetures & dependent feture/target put all independent fetures --> X & dependent feture/target --> Y

```python
X = insurence_data.drop(columns='charges', axis=1)
Y = insurence_data['charges']

print(Y)
print(X)
```

### spliting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
print(Y.shape, Y_test.shape, Y_train.shape)
```

### Model training

```python
# loading the linear regration model
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

### Model Evaluatiom

```python
# predection on training data
training_data_prediction = regressor.predict(X_train)
```

```python
r2_train = metrics.r2_score(Y_train, training_data_prediction )
print('R Squ value', r2_train)
```

```python
# predection on test data
test_data_prediction = regressor.predict(X_test)
```

```python
r2_test = metrics.r2_score(Y_test, test_data_prediction )
print('R Squ value', r2_test)
```

```python


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Extract features
X_age = X['age'].values
X_bmi = X['bmi'].values
y = Y.values

# Create grid for surface
age_range = np.linspace(X_age.min(), X_age.max(), 50)
bmi_range = np.linspace(X_bmi.min(), X_bmi.max(), 50)
age_grid, bmi_grid = np.meshgrid(age_range, bmi_range)

# Prepare input for prediction
X_constant = np.mean(X, axis=0)                   # avg of all features
X_input = np.tile(X_constant, (age_grid.size, 1)) # duplicate row
X_input[:, 0] = age_grid.ravel()                  # replace Age col
X_input[:, 2] = bmi_grid.ravel()                  # replace BMI col

# Predict insurance charges
y_pred_grid = regressor.predict(X_input)
y_pred_grid = y_pred_grid.reshape(age_grid.shape)

# Plot 3D scatter + regression surface
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection="3d")

# Scatter actual points
ax.scatter(X_age, X_bmi, y, 
           c="blue", s=40, alpha=0.6, edgecolor="k", label="Actual Data")

# Regression surface with colormap
surf = ax.plot_surface(age_grid, bmi_grid, y_pred_grid, 
                       cmap="viridis", edgecolor="none", alpha=0.6)

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=10, label="Predicted Insurance Cost")

# Labels & title
ax.set_xlabel("Age", labelpad=12, fontsize=11)
ax.set_ylabel("BMI", labelpad=12, fontsize=11)
ax.set_zlabel("Insurance Cost", labelpad=12, fontsize=11)
ax.set_title("3D Regression: Age & BMI vs Insurance Cost", fontsize=14, fontweight="bold")

# Tweak view angle
ax.view_init(elev=25, azim=120)

# Show legend
ax.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# 2D Visualization: BMI vs Insurance Cost

# Select BMI column (index 2 in your dataset)
X_bmi = X['bmi'].values
y = Y.values

# Sort values for cleaner plotting
sorted_idx = np.argsort(X_bmi)
X_bmi_sorted = X_bmi[sorted_idx]
y_sorted = y[sorted_idx]

# Create input with BMI range, keeping other features constant (mean values)
bmi_range = np.linspace(X_bmi.min(), X_bmi.max(), 100).reshape(-1, 1)
X_constant = np.mean(X, axis=0)
X_input = np.tile(X_constant, (100, 1))
X_input[:, 2] = bmi_range.ravel()

y_pred_line = regressor.predict(X_input)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(X_bmi, y, alpha=0.4, color="blue", label="Actual data")
plt.plot(bmi_range, y_pred_line, color="red", linewidth=2, label="Regression line")
plt.xlabel("BMI")
plt.ylabel("Insurance Cost (charges)")
plt.title("Linear Regression: BMI vs Insurance Cost")
plt.legend()
plt.show()
```
